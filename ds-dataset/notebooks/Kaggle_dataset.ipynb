### Cleaning Data

```python
def bersihkan_teks(teks):
    if pd.isna(teks):
        return ""
    teks = str(teks).lower() # Mengubah teks menjadi huruf kecil
    teks = re.sub(r'http\S+|www\S+|https\S+', '', teks, flags=re.MULTILINE) # Menghapus link URL
    teks = re.sub(r'[^\w\s]', ' ', teks) # Menghapus tanda baca
    teks = re.sub(r'\s+', ' ', teks).strip() # Menghapus spasi berlebih
    return teks

# Mengaplikasikan fungsi pembersihan ke kolom yang akan dianalisis
df['clean_tags'] = df['all_tags'].apply(bersihkan_teks)
df['clean_judul'] = df['Judul'].apply(bersihkan_teks)
df['clean_konten'] = df['Content'].apply(bersihkan_teks)
```

```python
def klasifikasi_berita(row):
    # Urutan prioritas: Tags -> Judul -> Isi Konten
    teks_untuk_dicek = [row['clean_tags'], row['clean_judul'], row['clean_konten']]

    for teks in teks_untuk_dicek:
        if teks: # Jika teksnya tidak kosong
            for kategori, kata_kunci in kamus_kategori.items():
                # Jika ada salah satu kata kunci yang cocok di dalam teks
                if any(kunci in teks for kunci in kata_kunci):
                    return kategori

    return 'Lainnya' # Jika tidak ada yang cocok sama sekali
```

```python
df['kategori_isu'] = df.apply(klasifikasi_berita, axis=1)

print("Proses pembersihan dan klasifikasi selesai!")
```

```python
def bersihkan_artefak_redaksi(teks):
    """
    Menghapus penanda identitas media di awal teks seperti:
    - "Jakarta, Kompas.com - "
    - "ANTARA News - "
    - "Reporter: Nama - "
    - URL dan tanda baca standar (sudah ditangani bersihkan_teks())
    """
    if pd.isna(teks) or teks == "":
        return ""

    # Hapus pola "Kota, Nama_Media - " atau "Nama_Media - " di awal
    teks = re.sub(
        r'^[\w\s]+,\s*[\w\s\.]+\s*[-–]\s*',  # "Jakarta, Kompas.com - "
        '', teks, flags=re.IGNORECASE
    )
    # Hapus pola nama media standalone di awal: "ANTARA - ", "detikcom - "
    teks = re.sub(
        r'^(antara(\s*news)?|kompas\.?com|detik\.?com|tribun\w*|cnnindonesia|tempo\.?co'
        r'|republika|liputan6|okezone)\s*[-–]\s*',
        '', teks, flags=re.IGNORECASE
    )
    # Hapus pola "Reporter: Nama Wartawan - "
    teks = re.sub(r'^reporter\s*:\s*.+?[-–]\s*', '', teks, flags=re.IGNORECASE)
    # Hapus URL yang mungkin tersisa
    teks = re.sub(r'http\S+|www\S+|https\S+', '', teks)
    # Rapikan spasi
    teks = re.sub(r'\s+', ' ', teks).strip()
    return teks

# Terapkan ke kolom konten yang sudah di-clean sebelumnya
df['teks_bersih'] = df['Content'].apply(bersihkan_artefak_redaksi)
# Lalu terapkan cleaning standar (lowercase, hapus tanda baca, spasi)
df['teks_bersih'] = df['teks_bersih'].apply(bersihkan_teks)

print("Contoh hasil pembersihan artefak redaksi:")
print(df[['Content', 'teks_bersih']].head(3).to_string())
```

### Standardisasi

```python
# STEP 1: Konversi tanggal DULU
def konversi_tanggal_berita(waktu):
    if pd.isna(waktu):
        return '2026-03-26'
    waktu = str(waktu).strip()
    bulan_map = {
        'Januari': 'January', 'Februari': 'February', 'Maret': 'March',
        'April': 'April', 'Mei': 'May', 'Juni': 'June',
        'Juli': 'July', 'Agustus': 'August', 'September': 'September',
        'Oktober': 'October', 'November': 'November', 'Desember': 'December'
    }
    for id_bulan, en_bulan in bulan_map.items():
        waktu = waktu.replace(id_bulan, en_bulan)
    waktu = re.sub(r'\s*\|.*$', '', waktu).strip()
    try:
        return pd.to_datetime(waktu, format='%d %B %Y').strftime('%Y-%m-%d')
    except:
        pass
    try:
        return pd.to_datetime(waktu, format='%d/%m/%Y').strftime('%Y-%m-%d')
    except:
        pass
    return '2026-03-26'

df['tanggal'] = df['Waktu'].apply(konversi_tanggal_berita)

# STEP 2: username
portal_mapping = {
    'kompas': '@kompascom', 'detik': '@detikcom', 'tribun': '@tribunews',
    'antara': '@antaranews', 'tempo': '@tempodotco', 'cnnindonesia': '@cnnindonesia',
    'republika': '@republikaonline', 'liputan6': '@liputan6dotcom', 'okezone': '@okezone',
}
def map_username(source):
    if pd.isna(source):
        return '@unknown'
    source_lower = str(source).lower()
    for key, val in portal_mapping.items():
        if key in source_lower:
            return val
    return f'@{source_lower.replace(" ", "")}'

df['username'] = df['source'].apply(map_username)

# STEP 3: Rakit df_berita_ready
df_berita_ready = pd.DataFrame({
    'sumber'         : 'berita_kaggle',
    'tanggal'        : df['tanggal'],
    'username'       : df['username'],
    'teks_bersih'    : df['teks_bersih'],
    'kategori_isu'   : df['kategori_isu'],
    'urgensi'        : 'Belum Dilabeli',
    'sentimen'       : 'Belum Dilabeli',
    'lokasi_insiden' : 'Belum diekstrak',
})

print("Shape DataFrame siap merge:", df_berita_ready.shape)
print(df_berita_ready['tanggal'].value_counts().head(10))
print(f"Imputasi: {(df_berita_ready['tanggal'] == '2026-03-26').sum()}")
```

### Sampling

```python
# ============================================================
# SAMPLING FINAL: 1000 per kategori
# - Kategori > 1000 : undersample (random)
# - Kategori < 1000 : oversample (duplikasi acak)
# ============================================================

TARGET = 1000

def sample_kategori(group):
    n = len(group)
    if n >= TARGET:
        # Undersample
        return group.sample(n=TARGET, random_state=42)
    else:
        # Oversample: duplikasi acak sampai 1000
        extra = group.sample(n=TARGET - n, replace=True, random_state=42)
        return pd.concat([group, extra])

df_sampled = (
    df_berita_ready
    .groupby('kategori_isu', group_keys=False)
    .apply(sample_kategori)
    .reset_index(drop=True)
)

print("Distribusi setelah sampling:")
print(df_sampled['kategori_isu'].value_counts())
print(f"\nTotal data: {len(df_sampled)} baris")
```
